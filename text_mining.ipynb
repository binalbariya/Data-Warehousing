```python
#importing the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#used to get the HTML of the website that we are scraping the table off of
url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Tamil_Nadu'
requests.get(url)
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'lxml')
```

```python
#this soup contains all the HTML from the website
print(soup.prettify())
```

```python
table_data = soup.find('div', class_ = "barbox tright")
```

```python
table_data
```

```python
headers = ["Date","# Of Cases","# Of Deaths"]
```

```python
headers
```

```python
df = pd.DataFrame(columns = headers)
```

```python
print(table_data.find_all('tr'))
```

```python
for j in table_data.find_all('tr'):
        row_data = j.find_all('td')
```

```python
row_data
```

```python
row = [tr.text.strip() for tr in row_data]
```

```python
length = len(df)
df.loc[length] = row
```

```python
import requests
import lxml.html as lh
import pandas as pd
```

```python
url='http://pokemondb.net/pokedex/all'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
#Check the length of the first 20 rows
[len(T) for T in tr_elements[:20]]
```

```python
tr_elements = doc.xpath('//tr')
#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    print(i,name)
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 10, the //tr data is not from our table 
    if len(T)!=10:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
[len(C) for (title,C) in col]
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
```
